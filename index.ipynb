## Skin Cancer

After a physical exam, a doctor observes a blemish on a client's arm. The doctor is concerned that the blemish could be cancerous, but tells the patient to be calm and that it's probably benign. Of those with skin cancer, 100% have such blemishes. However, 20% of those without skin cancer also have such blemishes. If 15% of the population has skin cancer, what's the probability that this patient has skin cancer? 

> Hint: Be sure to calculate the overall rate of blemishes across the entire population.

```python
# Your code here
# P(a) = 0.15 (15% of the population has skin cancer)
# P(b|a) = 1 (all patients with skin cancer have a blemish)
# P(b not a) = 0.2 (20% of patients without skin cancer have a blemish)

# Prior probabilities
P_A = 0.15  # Probability of having skin cancer
P_not_A = 1 - P_A  # Probability of not having skin cancer

# Conditional probabilities
P_B_given_A = 1  # Probability of having a blemish given skin cancer
P_B_given_not_A = 0.2  # Probability of having a blemish given no skin cancer

# Total probability of having a blemish
P_B = P_B_given_A * P_A + P_B_given_not_A * P_not_A

# Applying Bayes' theorem
P_A_given_B = P_B_given_A * P_A / P_B

print(P_A_given_B) # Output: 0.46875
```

## Children

A couple has two children, the older of which is a boy. What is the probability that they have two boys?

```python
# Your solution P(2boys|older child is a boy)

# Prior probabilities
P_A = 1/4  # Probability of having two boys
P_not_A = 3/4  # Probability of not having two boys (i.e., having 1 boy and 1 girl or 2 girls)

# Conditional probabilities
P_B_given_A = 1  # Probability of having at least one boy given two boys
P_B_given_not_A = 1/3  # Probability of having at least one boy given not two boys

# Total probability of having at least one boy
P_B = P_B_given_A * P_A + P_B_given_not_A * P_not_A

# Applying Bayes' theorem
P_A_given_B = P_B_given_A * P_A / P_B

print(P_A_given_B)  # Output: 0.5
```

## A diagnostic test

A diagnostic test is advertised as being 99% accurate 

* If a patient has the disease, they  will test positive 99% of the time 

* If they don't have the disease, they will test negative 99% of the time  

* 1% of all people have this disease 

If a patient tests positive, what is the probability that they actually have the disease?

```python
# Your solution P(Disease | positive test)
def bayes(P_D, P_T_given_D, P_T_given_not_D):
    # calculate P(not D)
    P_not_D = 1 - P_D
    # calculate P(T)
    P_T = P_T_given_D * P_D + P_T_given_not_D * P_not_D
    # calculate P(D|T) using Bayes' theorem
    P_D_given_T = P_T_given_D * P_D / P_T
    return P_D_given_T

# probability of having the disease
P_D = 0.01
# probability of testing positive given the disease
P_T_given_D = 0.99
# probability of testing positive given not having the disease
P_T_given_not_D = 0.01

# calculate probability of having the disease given testing positive
result = bayes(P_D, P_T_given_D, P_T_given_not_D)
print(result) # Output is 0.5
```
